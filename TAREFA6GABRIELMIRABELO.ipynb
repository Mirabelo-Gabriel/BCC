1) A tensão elétrica medida em um circuito, em volts, tem a seguinte expressão no instante t, em segundos:
 v(t)=e−0.5tcos(2π0.8t)
Faça o gráfico da tensão em função do tempo.
Sugestão: fazer o gráfico com tempo entre 0 e 10 s com incremento de 0.05 s.

```python
t= np.arange(0 , 10 , 0.05) #Tempo medido de 0 a 10 a cada 0.05 seg.
t
```

```python
v= (np.exp(-0.5*t))* (np.cos(2*(np.pi*(0.8*t)))) #Resolução da formula
```

```python
plt.figure(figsize=(15, 7)) #tamanho da figura
plt.plot(t , v , marker='d', color='purple') #variáveis , pontos e cor do gráfico.
plt.xlabel('TEMPO', fontsize=25) #Titulo do eixo X
plt.ylabel('VOLTS', fontsize=25) #Titulo do eixo Y
plt.grid() #malha retangular
plt.show()

#OBS GRÁFICO SE A CORRENTE FOR ALTERNADA
```

```python
plt.figure(figsize=(15, 7)) #tamanho da figura
plt.plot(t , (np.abs(v)), marker='d', color='purple') #variáveis , pontos e cor do gráfico.
plt.xlabel('TEMPO', fontsize=25) #Titulo do eixo X
plt.ylabel('VOLTS', fontsize=25) #Titulo do eixo Y
plt.grid() #malha retangular
plt.show()

#OBS GRÁFICO SE A CORRENTE A SER MEDIDA FOR CONTÍNUA
```

Faça um gráfico da equivalência da temperatura em Fahrenheits em função da temperatura em Celsius:
F=9/5C+32
Faça o gráfico tomando 100 pontos entre -20 ∘C e 100 ∘C.

```python
c=np.linspace(-20 , 100 , 100)
c
```

```python
F= (9/5)*c + 32
plt.figure(figsize=(15, 7)) #tamanho da figura
plt.plot(c , F , marker='d', color='red') #variáveis , pontos e cor do gráfico.
plt.xlabel('Graus celsius', fontsize=25) #Titulo do eixo X
plt.ylabel('Graus Farenheit', fontsize=25) #Titulo do eixo Y
plt.grid() #malha retangular
plt.show()
```

3) Entre os anos de 1994 e 2022, a população de uma cidade cresce aproximadamente de acordo com a seguinte função:
P(a)=800000e^a−1994/40
em que a é o ano e P é o número de habitantes.

Faça o gráfico do número de habitantes entre 1994 e 2022.

```python
a= np.arange (1994, 2022, 1)
P= 800000*(np.exp((a -1994)/40))

plt.figure(figsize=(15, 7)) #tamanho da figura
plt.plot(a, P , marker='d', color='green') #variáveis , pontos e cor do gráfico.
plt.xlabel('Relação ano', fontsize=25) #Titulo do eixo x
plt.ylabel('Relação População', fontsize=25) #Titulo do eixo Y
plt.grid() #malha retangular
plt.show()
```
